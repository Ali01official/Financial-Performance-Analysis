Imporing csv file from data set

```python
import pandas as pd
fd=pd.read_csv(r"D:\Machine Learning\Dummy data practice\1. dummy data client wise revenue and expense, receivables and payables\01. financial performance data\Data\1. client wise financial performance data.csv")
print(fd)
```

Calculating gross profit margin ratio

```python
gross_profit_ratio=fd["gross profit"]/fd["revenue"]
print(gross_profit_ratio)
```

Calculating net profit margin ratio

```python
net_profit_ratio=fd["net profit"]/fd["revenue"]
print(net_profit_ratio)
```

adding gross profit and net profit ratios coloum in main file

```python
fd["gross profit ratio"]=gross_profit_ratio
fd["net profit ratio"]=net_profit_ratio
```

```python
fd.columns
```

Exporting data to a csv file

```python
fd.to_csv("client wise financial data.02.csv",index=False)
```

Calculation of Operation cost ratio

```python
fd_opratio=fd['operational cost']/fd['revenue']
print(fd_opratio)
```

Calculating admin expense ratio

```python
fd_adminratio=fd["admin expense"]/fd["revenue"]
print(fd_adminratio)
```

Determining monthly client wise revenue trend

```python
fd_revenuemonthwise=fd.groupby('month')["revenue"].sum()
print(fd_revenuemonthwise)
```

```python
fd_client_wise_monthly_trend=fd.groupby(["month","client name"])["revenue"].sum().reset_index()
print(fd_client_wise_monthly_trend)
```

Determining Client wise monthly net profit trend

```python
fd_client_wise_monthly_nptrend=fd.groupby(["month","client name"])["net profit"].sum().reset_index()
print(fd_client_wise_monthly_nptrend)

```

Determining month wise average net profit ratio

```python
fd_best_np_trend=fd.groupby("month")["net profit ratio"].mean().reset_index()
print(fd_best_np_trend)
```
